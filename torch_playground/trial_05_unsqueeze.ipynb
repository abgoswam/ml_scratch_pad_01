```python
import torch
```

```python
# Level 1: Basic tensor shapes
x = torch.randn(3, 4, 5)
print(f"Tensor x shape: {x.shape}")
print(f"Number of dimensions: {x.ndim}")
print(f"Tensor:\n{x}")
```

```python
# You can also use negative indices
x_unsqueeze_neg1 = x.unsqueeze(-1)
print(f"After unsqueeze(-1): {x_unsqueeze_neg1.shape}")
# print(f"Tensor:\n{x_unsqueeze_neg1}")

# You can also use negative indices
x_unsqueeze_neg2 = x[..., None]
print(f"After unsqueeze(-1): {x_unsqueeze_neg2.shape}")
# print(f"Tensor:\n{x_unsqueeze_neg2}")
```
