```python
import torch
import torch.nn as nn
```

```python
m = nn.ReLU()
```

```python
input = torch.randn(2).unsqueeze(0)

print(input)
print(input.shape)
```

```python
print(m(input))
print(m(-input))
```

```python
output = torch.cat((m(input), m(-input)))
print(output)
```

```python
output2 = torch.cat((m(input), m(-input)), dim=0)
print(output2)
```
