```python
import torch
import numpy as np
from torchvision import transforms

# Simulate raw PIL image pixels (0-255 range)
raw_pixels = np.random.randint(0, 256, size=(28, 28), dtype=np.uint8)

# Calculate mean and std BEFORE ToTensor
raw_mean = raw_pixels.mean()
raw_std = raw_pixels.std()

print("BEFORE ToTensor() - Raw PIL pixels (0-255):")
print(f"  Mean: {raw_mean:.2f}")
print(f"  Std:  {raw_std:.2f}")
print()

# Apply ToTensor (divides by 255)
to_tensor = transforms.ToTensor()
tensor = to_tensor(raw_pixels)

# Calculate mean and std AFTER ToTensor
tensor_mean = tensor.mean()
tensor_std = tensor.std()

print("AFTER ToTensor() - Tensor pixels (0-1):")
print(f"  Mean: {tensor_mean:.6f}")
print(f"  Std:  {tensor_std:.6f}")
print()

# Verify the scaling relationship
print("Verification (should be â‰ˆ 255):")
print(f"  raw_mean / tensor_mean = {raw_mean / tensor_mean.item():.2f}")
print(f"  raw_std / tensor_std   = {raw_std / tensor_std.item():.2f}")
```
